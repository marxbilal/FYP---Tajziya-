```python
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
def preproccess(text):
    ptext = ""
    ptext = remove_punctuation(text)
    ptext = replace_urls(ptext)
    ptext = replace_emails(ptext)
    ptext = replace_numbers(ptext)
    ptext = remove_english_alphabets(ptext)
#   normalize_whitespace flips text sequence
    ptext = normalize_whitespace(ptext)
#     ptext = deEmojify(ptext)
    ptext = normalize(ptext)
    return ptext
    
```
